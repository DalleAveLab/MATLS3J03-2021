```python
import statsmodels.graphics.gofplots as sm
import numpy as np
mean = np.mean(plant_growth)
std = np.std(plant_growth, ddof=1)
print(std)
plant_growth = plant_growth.sort_values('Seedling Height')
sm.qqplot(plant_growth,loc=mean['Seedling Height'],scale=float(std),line='45')
```

```python
from scipy.stats import norm
z = (18-mean)/std
print(z)
```

```python
pz = norm.cdf(z)
print(1 - pz)
```

```python
from scipy.stats import t
Ct = t.ppf(0.975,60)
print(Ct)
```

```python
LB = mean - Ct*std/np.sqrt(61)
UB = mean + Ct*std/np.sqrt(61)
print(str(LB) + "<= mu <= " + str(UB) )
```
